```python
import numpy as np
import pandas as pd

import warnings

warnings.filterwarnings('ignore')
```

# Load Actual Data

```python
prod_df = pd.read_csv('prod-region.csv', sep=';')
```

```python
prod_df.head()
```

```python
#reduce information
prod_df.drop(['Région', 'Code région', 'Plage de puissance injection', 'Filière de production', 'Nb points injection', 'Courbe Moyenne n°1 (Wh)', 'Indice représentativité Courbe n°1 (%)',
         'Courbe Moyenne n°2 (Wh)', 'Indice représentativité Courbe n°2 (%)',
         'Indice représentativité Courbe n°1 + n°2 (%)', 'Jour max du mois (0/1)', 'Semaine max du mois (0/1)', 'Courbe Moyenne n°1 + n°2 (Wh)'],
        axis='columns', inplace=True)
prod_df.dropna(inplace=True)
prod_df['Horodate'] = pd.to_datetime(prod_df['Horodate'].values, utc=True)
prod_df['Horodate'] = prod_df['Horodate'].dt.tz_localize(None)
prod_df = prod_df[prod_df['Horodate'].dt.month != 3]
prod_df = prod_df.groupby('Horodate').sum().iloc[::-1]
prod_df = prod_df.reset_index()
prod_df.head(100)
```

# Load Forecasted Data

```python
fcst_df = pd.read_csv('sampleSolarT12022.csv', sep=',')
```

```python
fcst_df.head()
```

```python
fcst_df.drop(['Nb points injection', 'Courbe Moyenne n°1 + n°2 (Wh)', 'Filière de production'], axis='columns', inplace=True)
fcst_df['Horodate'] = pd.to_datetime(fcst_df['Horodate'].values, utc=True)
fcst_df['Horodate'] = fcst_df['Horodate'].dt.tz_localize(None)
fcst_df = fcst_df[fcst_df['Horodate'].dt.month != 3]
fcst_df = fcst_df.groupby('Horodate').sum().iloc[::-1]
fcst_df = fcst_df.reset_index()
fcst_df.head(100)
```

# Comparison between Actual and Forecasted Data

```python
import plotly.graph_objects as go
```

```python
# plot fcst_df and prod_df on the same plotly figure
fig = go.Figure()
fig.add_trace(go.Scatter(x=prod_df['Horodate'], y=prod_df['Total énergie injectée (Wh)'], name='Actual'))
fig.add_trace(go.Scatter(x=fcst_df['Horodate'], y=fcst_df['Total énergie injectée (Wh)'], name='Forecast'))
fig.update_layout(
    title='Actual and Forecasted Production',
    xaxis_title='Date',
    yaxis_title='Production (Wh)'
)
```

```python
# Remove records that don't have the same timestamp in both datasets
_prod_df = prod_df[prod_df['Horodate'].isin(fcst_df['Horodate'])]
# sort the dataframe by date
_prod_df = _prod_df.sort_values(by='Horodate')
```

```python
_prod_df
```

```python
_fcst_df = fcst_df[fcst_df['Horodate'].isin(_prod_df['Horodate'])]
_fcst_df = _fcst_df.sort_values(by='Horodate')
```

```python
_fcst_df
```

```python
_prod_df = _prod_df[((_prod_df['Horodate'].dt.hour > 8) & (_prod_df['Horodate'].dt.hour < 16)) | ((_prod_df['Horodate'].dt.hour < 4) & (_prod_df['Horodate'].dt.hour > 22))]
_fcst_df = _fcst_df[((_fcst_df['Horodate'].dt.hour > 8) & (_fcst_df['Horodate'].dt.hour < 16)) | ((_fcst_df['Horodate'].dt.hour < 4) & (_fcst_df['Horodate'].dt.hour > 22))]
```

```python
fcst_arr, prod_arr = _fcst_df['Total énergie injectée (Wh)'].values, _prod_df['Total énergie injectée (Wh)'].values
```

```python
print("The mean percentage difference over the forecasted data is: {:.2%}".format(np.abs(np.mean((fcst_arr - prod_arr) / prod_arr))))
```
