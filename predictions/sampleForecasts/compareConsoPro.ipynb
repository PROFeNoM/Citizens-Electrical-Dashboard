```python
import numpy as np
import pandas as pd

import warnings

warnings.filterwarnings('ignore')
```

# Load Actual Pro Data

```python
cons_df = pd.read_csv('conso-inf36-region.csv', sep=';')
```

```python
cons_df
```

```python
cons_df.drop(['Région', 'Code région', 'Plage de puissance souscrite', 'Nb points soutirage', 'Courbe Moyenne n°1 (Wh)', 'Indice représentativité Courbe n°1 (%)', 'Courbe Moyenne n°2 (Wh)', 'Indice représentativité Courbe n°2 (%)', 'Indice représentativité Courbe n°1 + n°2 (%)', 'Jour max du mois (0/1)', 'Semaine max du mois (0/1)', 'Courbe Moyenne n°1 + n°2 (Wh)'], axis='columns', inplace=True)
cons_df.dropna(inplace=True)
cons_df['Horodate'] = pd.to_datetime(cons_df['Horodate'].values, utc=True)
cons_df['Horodate'] = cons_df['Horodate'].dt.tz_localize(None)
cons_df = cons_df[cons_df['Profil'].str.contains("PRO")]
cons_df = cons_df.groupby('Horodate').sum().iloc[::-1]
cons_df.head(100)
```

# Load Forecasted Pro Data

```python
fcst_df = pd.read_csv('sampleProT12022.csv', sep=',')
```

```python
fcst_df
```

```python
fcst_df.drop(['Courbe Moyenne n°1 + n°2 (Wh)', 'Nb points soutirage', 'Profil'], axis='columns', inplace=True)
fcst_df['Horodate'] = pd.to_datetime(fcst_df['Horodate'].values, utc=True)
fcst_df['Horodate'] = fcst_df['Horodate'].dt.tz_localize(None)
fcst_df.set_index('Horodate', inplace=True)
fcst_df.head(100)
```

# Comparison between Actual and Forecasted Data

```python
import plotly.graph_objects as go
```

```python
# plot fcst_df and prod_df on the same plotly figure
fig = go.Figure()
fig.add_trace(go.Scatter(x=cons_df.index, y=cons_df['Total énergie soutirée (Wh)'], name='Actual'))
fig.add_trace(go.Scatter(x=fcst_df.index, y=fcst_df['Total énergie soutirée (Wh)'], name='Forecast'))
fig.update_layout(
    title='Actual and Forecasted Consumption',
    xaxis_title='Date',
    yaxis_title='Consumption (Wh)'
)
```

```python
# Remove records that don't have the same timestamp in both datasets
_cons_df = cons_df[cons_df.index.isin(fcst_df.index)]
```

```python
_cons_df
```

```python
_fcst_df = fcst_df[fcst_df.index.isin(_cons_df.index)]
```

```python
_fcst_df
```

```python
fcst_arr = _fcst_df['Total énergie soutirée (Wh)'].values
cons_arr = _cons_df['Total énergie soutirée (Wh)'].values
```

```python
print("The mean percentage difference over the forecasted data is: {:.2%}".format(np.abs(np.mean((fcst_arr - cons_arr) / cons_arr))))
```
